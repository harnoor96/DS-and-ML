```python
from keras.models import load_model
import cv2
import numpy as np
import os
from skimage import io
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import tensorflow as tf
tf.get_logger().setLevel('INFO')
```

```python
frames = 16
rows = 120
cols = 120
channels = 3
```

```python
def init_batch_data():
    batch_data = np.zeros((1, frames, rows, cols, channels))
    return batch_data
```

```python
folder_names = ['Right Swipe', 'Thumbs Down', 'Stop']
```

```python
def showImage(array, idx, folder_name):
    a = array[idx]
    image = io.imread(folder_name + '/' + a)
    plt.subplot(2, 2,1)
    plt.imshow(image)
    plt.show()
```

```python
class_names = ['Left Swipe', 'Right Swipe', 'Stop', 'Thumbs down', 'Thumbs up']
model = load_model('Gesture_Trained_Model.h5')
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

def getPredictedGesture(images_data):
    classes = np.argmax(model.predict(images_data), axis = -1)
    names = [class_names[i] for i in classes]

    print('Predicted gesture: ')
    print(names)
```

```python
for folder in folder_names: 
    images_data = init_batch_data()
    imgs = sorted(os.listdir(folder))
    imgs = imgs[7:23]
    print('Actual gesture: ' + folder)
    for idx, img in enumerate(imgs):
        image = cv2.imread(folder + '/' + img, cv2.IMREAD_COLOR)
        if (image.shape[0] != image.shape[1]):
            image=image[0:120,20:140]
        resized = cv2.resize(image, (120, 120), interpolation = cv2.INTER_AREA)
        images_data[0, idx] = (resized)
        if idx % 6 == 0:
            showImage(imgs, idx, folder)
    getPredictedGesture(images_data)
    print()
```
